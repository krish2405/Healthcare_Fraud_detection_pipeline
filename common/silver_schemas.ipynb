```python
def create_empty_table(table_name, schema):
    """
    Create an empty Delta table using the supplied schema.
    """

    empty_df = spark.createDataFrame([], schema)

    (
        empty_df.write
        .format("delta")
        .mode("overwrite")
        .saveAsTable(table_name)
    )
```

```python
from pyspark.sql.types import StructType, StructField, StringType, TimestampType

def create_reject_schema(schema):
    reject_fields=schema.fields.copy()
    reject_fields.append(
        StructField("reject_reason", StringType(), True)
    )

    reject_fields.append(
        StructField("reject_ts", TimestampType(), True)
    )

    return StructType(reject_fields)
```
