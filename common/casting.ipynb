```python
SCHEMA_MAPPING = {
    "beneficiary": beneficiary_schema,
    "provider": provider_schema,
    "inpatient": inpatient_schema,
    "outpatient": outpatient_schema
}
```

```python
from pyspark.sql.functions import col

def cast_to_schema(df,source_name):

    print(f"casting columns ........")

    if source_name not in SCHEMA_MAPPING:
        raise ValueError(f"Unknown source name: {source_name}. Check your SCHEMA_MAPPING.")
        
    schema = SCHEMA_MAPPING[source_name]
    print(f"Casting columns based on schema for: {source_name}")


    print(f"casting columns ........ based on {schema}")
    
    for field in schema.fields:
        column_name =field.name
        if column_name not in df.columns:
            continue
        if column_name =='ingestion_ts':
            continue
        
        print(f"{field.name} -> {field.dataType}")
        df=df.withColumn(
            column_name,
            col(column_name).cast(field.dataType)
        )
    return df

```
