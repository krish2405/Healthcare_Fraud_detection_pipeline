```python
%sql
create CATALOG if not exists healthcare_catalog;

create schema if not exists healthcare_catalog.bronze; 

create schema if not exists healthcare_catalog.silver; 

create schema if not exists healthcare_catalog.gold; 

create schema if not exists healthcare_catalog.raw; 

```

```python
%sql
create volume if not exists healthcare_catalog.raw.source
```

```python
%sql
create schema if not exists healthcare_catalog.metadata
```

```python
%sql
CREATE TABLE healthcare_catalog.metadata.file_tracker (
    source_name STRING,
    file_name STRING,
    file_path STRING,
    processed_ts TIMESTAMP,
    status STRING
)
USING DELTA;
```

```python
%sql
CREATE TABLE healthcare_catalog.metadata.ingestion_log
(
    source_name STRING,
    table_name STRING,
    file_name STRING,
    file_path STRING,
    status STRING,
    records_loaded BIGINT,
    error_message STRING,
    ingestion_ts TIMESTAMP
)
USING DELTA;
```

```python
%sql
ALTER TABLE healthcare_catalog.metadata.ingestion_log
ADD COLUMNS (
    start_ts TIMESTAMP,
    end_ts TIMESTAMP,
    duration_seconds DOUBLE
);
```
