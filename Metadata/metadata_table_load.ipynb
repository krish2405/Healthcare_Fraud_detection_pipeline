```python
import configparser

config=configparser.ConfigParser()
config.read('/Workspace/Users/krishraj2408@gmail.com/Healthcare_Fraud_detection_pipeline/config2.ini')
```

```python
folders=dbutils.fs.ls(config['VOLUME']['volume_path'])

print(folders)
```

```python
from pyspark.sql.functions import current_timestamp, lit

rows = []
for folder in folders:
    files = dbutils.fs.ls(folder.path)
    for file in files:
        rows.append({
            "source_name": folder.name,
            "file_name": file.name,
            "file_path": file.path,
            "status": "processed"
        })

metadata_df = spark.createDataFrame(rows)
metadata_df = metadata_df.withColumn("processed_ts", current_timestamp())
metadata_df.write.mode('append').saveAsTable('healthcare_catalog.metadata.file_tracker')
```
